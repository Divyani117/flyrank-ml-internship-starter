## Research Question

Which pages with meaningful search visibility have relatively low click-through rates and should be prioritized for review?

**Unit of analysis:** One page.

**Decision:** Decide which pages should be reviewed first for possible improvements to titles, descriptions, or search presentation.

**Output:** A ranked opportunity score or prioritized list of pages.

**Action:** An SEO or content team can review the highest-priority pages and decide whether their search presentation should be improved.

**Cost of a wrong recommendation:** A false positive could waste time reviewing a page that does not need improvement. A false negative could miss a page with a genuine opportunity to gain more clicks. Therefore, the output should support human review rather than automatically changing content.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

url = "https://raw.githubusercontent.com/Divyani117/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)

print("Number of pages:", len(df))
print("Median CTR:", df["ctr"].median())
print("Median impressions_90d:", df["impressions_90d"].median())
print("Median average position:", df["avg_position"].median())
```

The starter dataset contains 30,000 pages. The median click-through rate (CTR) is 0.07 (7%), the median number of impressions over 90 days is 731, and the median average search position is 10.8.

These numbers show that the dataset contains a substantial number of pages with measurable search visibility and engagement data. This makes the CTR / Engagement Opportunity Scoring lane worth investigating because pages with meaningful impressions but relatively low CTR may represent useful candidates for human review and prioritization.

These are descriptive observations from the starter dataset and do not prove that changing a page will increase clicks.
